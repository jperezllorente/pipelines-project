#### The first thing we want to check is the which are the most popular styles in general (including the four playlists). 

```python
sns.countplot(y="style", data=data, palette="icefire",
              order=data["style"].value_counts().iloc[:20].index)
```

```python
sns.scatterplot(x="percentage", y="style", hue="playlist", data=data)
```

As can be observed, the most popular music styles globally, according to Spotify users, are: 

    - Pop
    - Hip hop 
    - Latin
    - Reggaeton
    - Trap
    - Rap

It's somehow surprising that latin music is one of the most popular, as non-spanish speakers wont understand a single word, but at the same time, the reason for its popularity can be that it is one of the best music genres to dance and have a party. #### Next, we will check how the styles popularity haas changed from 2019 to 2020. In order to do this,  we will create a subdataframe that includes only the observations of GLB_top30_2020 and data(top 2019)

```python
year_ev = data[data["playlist"].isin(['GLB_top 2020', 'GLB_top 2019'])]
```

```python
sns.countplot(y="style", data = year_ev, hue = 'playlist' ,palette="icefire",
              order=data["style"].value_counts().iloc[:30].index)
```

In this plot we can see that pop music is the most popular music style in both years. This is not surprising as it is the most commecial genre, and the one that can be most enjoyed by different people. 

It can be seen that hip hop has gained popularity in 2020 from 2019, and the rest of the styles have kept the same trend. It is important to point out that we have only been able to use information of 30 songs form playlists of Spotify, what can distort the results.
